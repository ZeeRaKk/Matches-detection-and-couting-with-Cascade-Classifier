#### Importation des bibiliothèques

```python
import cv2 # Importation de la bibliothèque d'OpenCV

import matplotlib.pyplot as plt # Importation d'une bibliothèque de matplotlib permettant de plot les images dans le notebook
```

## Fonction Main

```python
MatchesCascade = cv2.CascadeClassifier('Matches_detection.xml') # Entrainement de notre classifieur sur notre fichier XML
def Main(name):
    image = cv2.imread(name) # Lecture de L'image
    image =cv2.resize(image,dsize=(1490,800)) # Réduction de la taille de l'image en gardant les mêmes proportion pour que la détection se fasse plus rapidement

    Matches = MatchesCascade.detectMultiScale(image,1.2) # Détection des allumettes sur notre image

    for(x, y, w, h) in Matches:
        cv2.rectangle(image, (x,y), (x+w, y+h),(0, 255, 0), 5) #On trace les rectangles sur l'image
    cv2.imwrite('output_' + name,image) # On enregistre l'image
    plt.figure(figsize=(15,10)) # On plot l'image
    plt.title(name + f" à {len(Matches)} allumette(s)")
    plt.imshow(image[:,:,::-1]) # ::-1 car plt possède la norme RGB et openCV celle en BGR
```

## Plot des images 

```python
try:
    Main('overlapped.png')
    Main('cross.png')
    Main('x.png')
    Main('h_bright.png')
    Main('h.png')
except:
    print("Les images doivent être dans le même dossier que le fichier notebook, mettez les et réessayer (Attention que le nom des images correspondent)")
```
